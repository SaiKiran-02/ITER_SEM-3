```python
import pandas as pd 
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

## Q 07 An NGO has participated in a three-week cultural festival. Using Pandas, store the sales (in Rs) made day wise for every week in a CSV file named “FestSales.csv” as shown in the table below.
## Write a Python script to display the sales for the three weeks using a Line chart. It should have the following:

### i. Chart title as “Festival Sales Report”.
### ii. X - axis label as Days.
### iii. Y - axis label as “Sales in Rs”.
### iv. Line colours are red for week 1, blue for week 2 and brown for week 3.

```python
file = open('FestSales.csv', "w")
file.write("week1,week2,week3\n")
for i in range(10):
    rows = ""
    for i in range(3):
        rows = rows + str(r.randint(3000,7000))+','
    rows = rows[0:len(rows)-1]    
    file.write('\n' + rows)
file.close()
```

```python
file = open('FestSales.csv','r')
file.read()
```

```python
df = pd.read_csv('FestSales.csv')
```

```python
df.plot(title = "Festival Sales Report",xlabel = "Days",ylabel = "Sales in rs")
```

```python
plt.plot(df['week1'], color='red', label='week 1')
plt.plot(df['week2'], color='blue', label='week 2')
plt.plot(df['week3'], color='brown', label='week 3')
```
